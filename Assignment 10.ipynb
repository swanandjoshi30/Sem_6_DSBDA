```python
import seaborn as sbn
import matplotlib.pyplot as plt
```

```python
data = sbn.load_dataset("iris")    #to load dataset
```

```python
data.info()
```

```python
data.describe()
```

```python
data.head()    #first five records
```

```python
data.tail()    #last 5 records
```

```python
plt.figure(figsize=(5,3))
plt.title("Distribution Of Sepal Length")
plt.hist(data["sepal_length"])
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()


# Creates a histogram for sepal_length
# Sets title: "Distribution of Sepal Length"
# Labels x-axis: "Sepal Length" and y-axis: "Frequency"
# Displays the plot
```

```python
plt.figure(figsize=(5,3))
plt.title("Distribution of Sepal Width")
plt.hist(data["sepal_width"])
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
plt.show()

 
# Creates a histogram for sepal_width
# Sets title: "Distribution of Sepal Width"
# Labels x-axis: "Sepal Width" and y-axis: "Frequency"
# Displays the plot
```

```python
plt.figure(figsize=(5,3))
plt.title("Distribution Of Petal Length")
plt.hist(data["petal_length"])
plt.xlabel = "Petal Length"
plt.ylabel = "Frequency"
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title("Distribution Of Petal Width")
plt.hist(data["petal_width"])
plt.xlabel = "{Petal} Width"
plt.ylabel = "Frequency"
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title("Distribution of Sepal Length")
sbn.boxplot(y="sepal_length",x="species",data=data,orient="v")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title("Distribution Sepal Width")
sbn.boxplot(y="sepal_width",x="species",data=data,orient="v")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title("Distribution Of Petal Length")
sbn.boxplot(y="petal_width",x="species",data=data,orient="v")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title("Distribution Of Petal Width")
sbn.boxplot(y="petal_width",x="species",data=data,orient='v')
plt.show()
```
