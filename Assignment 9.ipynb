```python
import seaborn as sbn
```

```python
data= sbn.load_dataset("titanic")
```

```python
data.head()
```

```python
data.tail()
```

```python
sbn.boxplot(y="age",x="sex",data=data,hue="survived")
```

Median Age : The median age appears similar for both males and females, but with some variation in survivors vs. non-survivors.

Age Distribution : The age range for males is wider, with more outliers, while female ages are more concentrated.

Survival Trend : Female survivors generally have a higher survival rate compared to males, reinforcing the "women and children first" policy.

Older Male Non-Survivors : Many older males did not survive, suggesting that survival rates were lower for older men.

Children's Survival : Younger age groups, especially among females, have a higher survival rate.
